```python
# Dependencies and Setup
import pandas as pd
import numpy as np
# File to Load (Remember to Change These)
file_to_load = "Resources/purchase_data.csv"

# Read Purchasing File and store into Pandas data frame
purchase_df = pd.read_csv(file_to_load)
purchase_df.head()
```

## Player Count

* Display the total number of players


```python
# Calculating the total number of players (excluding duplicates)
total_players = len(purchase_df["SN"].unique())
total_players

# Create data frame  
total_players_df = pd.DataFrame({
    "Total Players": [total_players]
})
total_players_df
```

## Purchasing Analysis (Total)

* Run basic calculations to obtain number of unique items, average price, etc.


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame


```python
# Calculate number of unique items 
unique_items = len(purchase_df["Item ID"].unique())

# Calculate Average Purchase Price 
avg_purchase_price = sum(purchase_df["Price"]) / len(purchase_df["Price"])

# Calculate number of purchases 
num_purchases = len(purchase_df["Purchase ID"].unique())

# Calculate Total Revenue
total_revenue = sum(purchase_df["Price"])
```

```python
# Create summary table for purchase analysis
summary_purchase_analysis = pd.DataFrame({
    "Number of Unique Items": [unique_items],
    "Average Price": [avg_purchase_price],
    "Number of Purchases": [num_purchases],
    "Total Revenue": [total_revenue]
    
})
```

```python
# Formatting 
summary_purchase_analysis["Average Price"] = summary_purchase_analysis["Average Price"].map("${:.2f}".format)
summary_purchase_analysis["Total Revenue"] = summary_purchase_analysis["Total Revenue"].map("${:,.2f}".format)
summary_purchase_analysis
```

## Gender Demographics

* Percentage and Count of Male Players


* Percentage and Count of Female Players


* Percentage and Count of Other / Non-Disclosed




```python
# Identify the different genders 
gender_list = list(purchase_df["Gender"].unique())

# Filter dataset to only male
male_data_df = purchase_df.loc[purchase_df["Gender"] == "Male", :]
male_data_df

# Calculate percentage and count of Male Players 
male_count = len(male_data_df["SN"].unique())
male_percentage = (male_count / total_players) * 100

# Filter dataset to only female
female_data_df = purchase_df.loc[purchase_df["Gender"] == "Female", :]
female_data_df

# Calculate percentage and count of Female Players 
female_count = len(female_data_df["SN"].unique())
female_percentage = (female_count / total_players) * 100

# Filter dataset to only Other / Non-Disclosed
other_data_df = purchase_df.loc[purchase_df["Gender"] == "Other / Non-Disclosed", :]
other_data_df

# Calculate percentage and count of Other / Non-Disclosed 
other_count = len(other_data_df["SN"].unique())
other_percentage = (other_count / total_players) * 100
```

```python
# Create summary table for gender demographics
summary_gender_demo = pd.DataFrame({
    "Gender": ["Male","Female","Other / Non-Disclosed"],
    "Total Count": [male_count, female_count, other_count],
    "Percentage of Players": [male_percentage, female_percentage, other_percentage] 
})

```

```python
# Formatting 
summary_gender_demo["Percentage of Players"] = summary_gender_demo["Percentage of Players"].map("{:.2f}%".format)
summary_gender_demo
```


## Purchasing Analysis (Gender)

* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. by gender




* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
# Purchase Count by Gender 
f_purchase_count = female_data_df["Purchase ID"].count()
m_purchase_count = male_data_df["Purchase ID"].count()
o_purchase_count = other_data_df["Purchase ID"].count()

# Average Purchase Price by Gender 
f_avg_price = female_data_df["Price"].mean()
m_avg_price = male_data_df["Price"].mean()
o_avg_price = other_data_df["Price"].mean()

# Total Purchase Value by Gender 
f_total_purchase = female_data_df["Price"].sum() 
m_total_purchase = male_data_df["Price"].sum() 
o_total_purchase = other_data_df["Price"].sum()

# Average Total Purchase per Person by Gender 
f_avg_purchase = f_total_purchase / female_count 
m_avg_purchase = m_total_purchase / male_count
o_avg_purchase = o_total_purchase / other_count
```

```python
# Create summary table for purchasing analysis by gender 
summary_purchase_analysis_gender = pd.DataFrame({
    "Gender": ["Female","Male","Other / Non-Disclosed"],
    "Purchase Count": [f_purchase_count, m_purchase_count , o_purchase_count],
    "Average Purchase Price": [f_avg_price, m_avg_price, o_avg_price],
    "Total Purchase Value": [f_total_purchase, m_total_purchase, o_total_purchase],
    "Avg Total Purchase Value per Person": [f_avg_purchase, m_avg_purchase, o_avg_purchase]
})
```

```python
# Formatting 
summary_purchase_analysis_gender["Average Purchase Price"] = summary_purchase_analysis_gender["Average Purchase Price"].map("${:.2f}".format)
summary_purchase_analysis_gender["Total Purchase Value"] = summary_purchase_analysis_gender["Total Purchase Value"].map("${:,.2f}".format)
summary_purchase_analysis_gender["Avg Total Purchase Value per Person"] = summary_purchase_analysis_gender["Avg Total Purchase Value per Person"].map("${:.2f}".format)
summary_purchase_analysis_gender
```

## Age Demographics

* Establish bins for ages


* Categorize the existing players using the age bins. Hint: use pd.cut()


* Calculate the numbers and percentages by age group


* Create a summary data frame to hold the results


* Optional: round the percentage column to two decimal points


* Display Age Demographics Table


```python
# Caluclate the min and max ages in dataset
min_age = purchase_df["Age"].min()
max_age = purchase_df["Age"].max()

# Identify the list of ages 
bins = [0, 9, 14, 19, 24, 29, 34, 39, 100]

# Create the names for the bins 
group_names = ["<10", "10-14", "15-19","20-24","25-29","30-34","35-39","40+"]

# Categorize existing players using age bin
purchase_df["Age Group"] = pd.cut(purchase_df["Age"], bins, labels = group_names)

# Group the purchase_df dataframe by the age demographics and 
# count number of *unique* players that fall within the bins 
age_group_df = purchase_df.groupby(["Age Group"])
age_count = purchase_df.groupby(["Age Group"])["SN"].nunique()

# Calculate percentages by age group
age_percent = (age_count / total_players) * 100
```

```python
# Create summary table for age demographics by gender 
summary_age_demo_gender = pd.DataFrame({
    "Total Count": age_count,
    "Percentage of Players": age_percent
})
```

```python
# Formatting 
summary_age_demo_gender["Percentage of Players"] = summary_age_demo_gender["Percentage of Players"].map("{:.2f}%".format)
summary_age_demo_gender
```

## Top Spenders

* Run basic calculations to obtain the results in the table below


* Create a summary data frame to hold the results


* Sort the total purchase value column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
# Group by "SN" to determine total spending per gamer 
gamer_spending_df = purchase_df.groupby(["SN"])

# Calculate purchase count by gamer 
gamer_purchase_count = gamer_spending_df["Purchase ID"].count()

# Calculate average purchase price for each gamer 
gamer_avg_price = gamer_spending_df["Price"].mean()

# Calculate total purchase value for each gamer 
gamer_total_value = gamer_spending_df["Price"].sum()

# Create dataframe that contains these values 
summary_top_spender = pd.DataFrame({
    "Purchase Count": gamer_purchase_count,
    "Average Purchase Price": gamer_avg_price,
    "Total Purchase Value": gamer_total_value,
    
})


# Sort values to determine top 5 spenders according to Total Purchase Value
top_spender_df = summary_top_spender.sort_values("Total Purchase Value", ascending=False)
```

```python
# Formatting 
top_spender_df["Average Purchase Price"] = top_spender_df["Average Purchase Price"].map("${:.2f}".format)
top_spender_df["Total Purchase Value"] = top_spender_df["Total Purchase Value"].map("${:.2f}".format)
top_spender_df.head(5)
```

## Most Popular Items

* Retrieve the Item ID, Item Name, and Item Price columns


* Group by Item ID and Item Name. Perform calculations to obtain purchase count, item price, and total purchase value


* Create a summary data frame to hold the results


* Sort the purchase count column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
# Groupby Item ID and Item Name 
popular_items_df = purchase_df.groupby(["Item ID", "Item Name"])

# Calculate Purchase Count by Item ID 
id_purchase_count = popular_items_df["Purchase ID"].count()

# Calculate Total Purchase Value 
id_total_value = popular_items_df["Price"].sum()

# Calculate Item Price 
id_price = id_total_value / id_purchase_count

# Create dataframe that contains these values 
summary_pop_item = pd.DataFrame({
    "Purchase Count": id_purchase_count,
    "Item Price": id_price,
    "Total Purchase Value": id_total_value
    
})

# Sort values to determine top 5 spenders according to Purchase Count 
pop_df = summary_pop_item.sort_values("Purchase Count", ascending=False)
```

```python
# Formatting 
pop_df["Item Price"] = pop_df["Item Price"].map("${:.2f}".format)
pop_df["Total Purchase Value"] = pop_df["Total Purchase Value"].map("${:.2f}".format)
pop_df.head(5)
```
